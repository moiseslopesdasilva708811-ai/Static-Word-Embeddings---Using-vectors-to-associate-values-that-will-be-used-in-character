## Static Word Embeddings - Using vectors to associate values ​​that will be used in character analysis and training

# Concept of Input Embedding

In natural language processing, the **Input Embedding** method transforms tokens through a tokenization process into vectors of real numbers. Each value within such a given vector represents the intensity of the specific semantic feature in an n-dimensional space.

### Semantic Geometry
Words that share meanings tend to point in similar directions. For example:
- The vectors for **"Dog"** and **"Cat"** will be close to each other.
- The vector for **"Computer"** will be far from **"Apple"**, unless the context is technology (Apple).

### Cosine Similarity

This is the most common metric used in Natural Language Processing (NLP). Instead of measuring the straight-line distance between two points, it measures the **cosine of the angle** between two vectors to determine how semantically related they are.

* **0° Angle:** The cosine is **1** (the words have identical orientation/synonyms).
* **90° Angle:** The cosine is **0** (the vectors are orthogonal and have no semantic relationship).
* **180° Angle:** The cosine is **-1** (the words have opposite meanings).

#### Mathematical Formula

Similarity is calculated by dividing the dot product of the vectors by the product of their norms (magnitudes):

$$\text{similarity} = \cos(\theta) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}$$

For a vector $\mathbf{A} = (a_1, a_2, \dots, a_n)$, the magnitude $\|\mathbf{A}\|$ is the square root of the dot product of the vector with itself:

$$\|\mathbf{A}\| = \sqrt{a_1^2 + a_2^2 + \dots + a_n^2}$$

Where:
* **$\mathbf{A} \cdot \mathbf{B}$**: Is the **dot product** between the vectors.
* **$\|\mathbf{A}\|$ and $\|\mathbf{B}\|$**: Represent the **Euclidean norm** (length) of each vector.

---

#### Python Implementation Example
To validate this calculation in a Jupyter notebook cell, you can use the `numpy` library:

```python
import numpy as np

# Defining vectors
A = np.array([3, 4])
B = np.array([4, 3])

# Vector magnitude using the norm rule
magnitude_A = np.linalg.norm(A)
magnitude_B = np.linalg.norm(B)

# Manual Cosine Similarity
dot_product = np.dot(A, B)
cosine_sim = dot_product / (magnitude_A * magnitude_B)

print(f"Magnitude of A: {magnitude_A}")
print(f"Cosine Similarity: {cosine_sim:.4f}")

## Core Calculations Based on Referenced Literature
# Applied Vector Algebra

Below is the correlation between fundamental concepts of Analytical Geometry and the architecture of modern neural networks.

### Direction Vector (Vetor Diretor)
* **In Winterle:** The vector that determines the direction and orientation of a line or plane.
* **In AI:** Represents the **direction of meaning**. In embedding space, the "angle" toward which the vector points defines the word's semantics. Words with similar directions share similar contexts.
* **Application:** Word Embeddings (Word2Vec, GloVe).

### Vector Magnitude / Norm (Módulo)
* **In Winterle:** Represents the magnitude or length of the vector: $\|\mathbf{v}\| = \sqrt{\sum a_i^2}$.
* **In AI:** Indicates the **intensity or importance** of a specific feature. In language models, normalized vectors (magnitude = 1) are often used to ensure word frequency doesn't distort similarity.
* **Application:** Layer Normalization (LayerNorm).

### Dot Product (Produto Escalar)
* **In Winterle:** A measure of the relationship between two vectors: $\mathbf{u} \cdot \mathbf{v}$.
* **In AI:** The **fundamental operation of neurons**. A neuron's calculation is essentially the dot product between weights ($W$) and inputs ($X$).
* **Application:** **Attention Mechanism** in Transformers (e.g., GPT). The model calculates the dot product between "Queries" and "Keys" to determine which words to "attend" to.

### Angle Between Vectors (Cosine Similarity)
* **In Winterle:** $\cos(\theta) = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \|\mathbf{v}\|}$.
* **In AI:** The primary metric for **Semantic Search**. If the cosine is close to 1, the model interprets the words as near-synonyms.
* **Application:** Recommendation systems and vector database searches.

---
*Reference: Based on "Álgebra Vetorial e Geometria Analítica" - Paulo Winterle.*

## Fine Tunning + RAG
